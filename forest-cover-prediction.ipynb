## 6. Data Preprocessing 

```python
# Removing Unnecessary column...
data.drop(["Id"], axis=1, inplace=True)
```

### 6.1 Defining x and y

```python
x=data.drop("Cover_Type", axis=1)
y=data["Cover_Type"]
```

### 6.2 Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## 7. Feature Engineering 

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train[num_cols]=scaler.fit_transform(x_train[num_cols])
x_test[num_cols]=scaler.transform(x_test[num_cols])
```

##### The numerical features were scaled using StandardScaler to normalize their ranges and improve model performance. Scaling was applied to all   numerical columns to ensure better convergence for algorithms like Logistic Regression, which are sensitive to feature magnitudes.

##### Tree-based models such as Decision Tree, Random Forest, and XGBoost are generally not affected by feature scaling. However, applying scaling does   not negatively impact their performance, so a single scaled dataset was used for simplicity and consistency across all models.
